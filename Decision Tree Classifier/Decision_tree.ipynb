```python
#Using small data sets--> iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris.data
```

```python
#Output feature -->
iris.target
```

```python
#Other way of loading iris data sets-->
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
#INDEPENDENT AND DEPENDENT FEATURES -->
X=df.iloc[:,:-1]
y=iris.target
```

```python
X
```

```python
y
```

```python
#Next step is to train test split -->
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
#importing decision tree classifier -->
from sklearn.tree import DecisionTreeClassifier
```

```python
#max_depth used for limiting decision tree further ## Also to reduce overfitting of the model-->##Post prunning -->
treemodel=DecisionTreeClassifier(max_depth=2)
```

```python
#POST PRUNING -->
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(treemodel,filled=True)
```

```python
y_pred=treemodel.predict(X_test)
```

```python
y_pred
```

```python
#To check the accuracy of the model -->
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test)) # Good model as 98% accuracy is achieved in both the metrics
```

```python
#Pre Punning-->
#Defining parameters-->
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2'],
    
}
```

```python
# Importing Grid search CV -->
from sklearn.model_selection import GridSearchCV
tree_model=DecisionTreeClassifier()
cv=GridSearchCV(tree_model,param_grid=parameter,cv=5,scoring='accuracy') #CV=5, 5 fold cross validation

```

```python
#Fitting the model into the cv-->
cv.fit(X_train,y_train)
```

```python
cv.best_params_   #Can be used to make the decision tree -->
```

```python
ypred=cv.predict(X_test)
ypred
```

```python
y_test
```

```python
score=accuracy_score(ypred,y_test)
print(score)
```

```python
print(classification_report(ypred,y_test))
```
