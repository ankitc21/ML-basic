```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris.values()
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data
```

```python
iris_data.info()
```

```python
iris_data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_Scaled=scaler.fit_transform(iris_data)
```

```python
X_Scaled
```

```python
X_Scaled.shape
```

```python
#Apply the PCA-->
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
pca
```

```python
pca_scaled=pca.fit_transform(X_Scaled)
```

```python
pca_scaled
```

```python
pca_scaled.shape
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
##Applying aglomerative clustering 
##To construct a dendogram
import scipy.cluster.hierarchy as sc
#plot the  dendogram -->
plt.figure(figsize=(20,7))
plt.title("Dendograms")

#Creation of dendogram -->
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.title('Dendogram')
plt.xlabel('Smaple Index')
plt.ylabel('Eucledian Distance')
```

```python
#Longest vertical line such that no horizontal line passes through it. -->
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,metric='euclidean',linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    agglo=AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='ward')
    agglo.fit(X_Scaled)
    score=silhouette_score(X_Scaled,agglo.labels_)
    silhouette_coefficients.append(score)
```

```python
#Plotting silhouette_score -->
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Clusteres")
plt.ylabel("silhouette_coefficients")
plt.show()
```
